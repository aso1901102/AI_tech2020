### openCVで画像を操作する

```
# 画像をネットからDL
import urllib.request as req
url = "https://raw.githubusercontent.com/masatokg/book-mlearn-gyomu/master/src/ch3/cv2io/test.jpg"
savefile = "test.jpg"
req.urlretrieve(url,savefile)

#　保存したファイルをopenCVで読み込む
import cv2
img = cv2.imread(savefile)
print(img)
```

あえて間違ったパスでimreadの戻り値を確認する

```
# 存在しないバスを指定して読み込みすると、戻り値はNone
ng_img = cv2.imread("nonfile")
print(ng_img)
```

### 読み込んだ画像データをインライン表示

```
# matplotlibのインライン表示を宣言
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
#ピクセル情報を変換してプロットを作る（BGR -> RGB)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

画像を保存する

```
# 保存するファイル名
out_png = "out.png"

#openCVの画像保存メソッド
cv2.imwrite(out_png,img)
```

### メモリ軸をオフにしたプロットを作って表示

```
# メモリ軸をオフにしたプロットを作る
plt.axis("off")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```
# 読み込んだ画像をリサイズ
#横幅600、縦幅300
im2 = cv2.resize(img,(600,300))

#リサイズした画像をファイルに保存
cv2.imwrite("out-resize.png",im2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

画像の切り取り（抽出）

```
#　切り取る領域をx,yの二次元配列を指定するように設定
# [開始行：終端行,開始列:終端行]

im3 = img[150:450,150:450]
plt.imshow(cv2.cvtColor(im3,cv2.COLOR_BGR2RGB))
plt.show()
```

```
 #　リサイズした画像を保存
 im4 = cv2.resize(im3,(400,400))
 cv2.imwrite("cut-resize.png",im4)
 plt.imshow(cv2.cvtColor(im3,cv2.COLOR_BGR2RGB))
 plt.show()
```
